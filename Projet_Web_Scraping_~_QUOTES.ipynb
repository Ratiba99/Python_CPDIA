**1. Définir le web scraping et les cas d’usage**
**Le web scraping** est une technique utilisée pour extraire des données de sites web. Il consiste à envoyer une requête HTTP pour récupérer le contenu d'une page, puis à analyser le HTML pour extraire des informations spécifiques.
**Cas d'usage :**
- Collecter des données pour des analyses de marché.
- Extraire des informations de sites de citations.
- Récupérer des avis produits ou des informations de tarifs.

2. Quels sont les étapes du web scraping ?
- Récupérer le contenu HTML de la page web.
- Analyser la structure HTML pour identifier les données.
- Extraire les informations nécessaires.
- Stocker les données sous une forme exploitable (par exemple, CSV ou Excel).

3. Faire une cartographie du site à analyser comme dans l’exemple suivant

Le site **quotes.toscrape.com** présente des pages de citations sous forme de blocs contenant :

- **Citation** : La citation est dans la balise `<span class="text">`.
- **Auteur** : L'auteur est dans la balise `<small class="author">`.
- **Tags** : Les tags sont dans des balises `<a class="tag">`.

### Table de correspondance :

| Données     | Chemin HTML                         |
|-------------|--------------------------------------|
| Citation    | `<span class="text">`               |
| Auteur      | `<small class="author">`            |
| Tags        | `<a class="tag">`                   |

5. Ecrire un script python pour récupérer l’ensemble des données de la page






```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL du site à scraper
url = "http://quotes.toscrape.com"

# Faire une requête GET pour récupérer le contenu de la page
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extraire les citations et auteurs
citations = soup.find_all('span', class_='text')
auteurs = soup.find_all('small', class_='author')

# Organiser les données
data = []
for citation, auteur in zip(citations, auteurs):
    data.append({
        "Citation": citation.text,
        "Auteur": auteur.text
    })

# L'affichage structuré
df = pd.DataFrame(data)
df.head()  # Afficher les 5 premières lignes
```
